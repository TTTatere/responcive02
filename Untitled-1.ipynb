```python
function filterGallery(category){

    const cards = document.querySelectorAll(".card");

    cards.forEach(card => {

        if(category === "all"){
            card.classList.remove("hide");
            return;
        }

        if(card.classList.contains(category)){
            card.classList.remove("hide");
        } else {
            card.classList.add("hide");
        }

    });
}
```
